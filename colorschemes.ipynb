```julia
# alternative search
function find_colorscheme(s::String)
    sorted_keys = sort([String(k) for k in keys(colorschemes)])
    return [Symbol(cs) for cs in sorted_keys if occursin(s, cs)]
end
```

```julia
# find colorscheme
find_colorscheme("seaborn")
```

```julia
 # inferno
 f(x, y) = cos(sqrt(x^2 + y^2)) / (1 + sqrt(x^2 + y^2))
 
heatmap(
    range(0, 2Ï€, length = 30),
    range(0, 2Ï€, length = 30),
    f,
    c = :inferno
    )
```

```julia
# viridis
x = range(0.25, 1.25, length=51)
y = range(-1.25, -0.25, length=51)
z = 4.0.*sin.(4*pi.*x)*cos.(6*pi.*reshape(y, 1, :))
contourf(x, y, z, c=:viridis, colorbar=true)
```
